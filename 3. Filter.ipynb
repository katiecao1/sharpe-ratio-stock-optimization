```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('SP.csv', index_col=1 ,parse_dates=True)
df.drop(labels='Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
#Calculate the daily return
df1 = df.pct_change()[1:]
df1.head()
```

# Filter 1: Average Daily Return

```python
#Get average daily return
df_mean = df1.mean(axis=0)
df_mean1 = []
    #Pick the top 10 stock with highest average daily returns
df_mean = df_mean.nlargest(10)
    #Append the top 10 stocks with their daily returns to a list
df_mean1 = df[df_mean.index]
df_mean1.head()
```

# Filter 2: Standard Deviation of Daily Return

```python
#Get standard deviation of each stock's daily return
df_std = df1.std()
df_std1 = []
#Pick 10 stocks with lowest variance, assuming least volatility
df_std = df_std.nsmallest(10)
#Append result to a list

df_std1 = df[df_std.index]
df_std1.head()
```

# Filter 3: Kurtosis and Skewness
 -0.8 to 0.8 for skewness and -3.0 to 3.0 for kurtosis are classified as normal distribution

```python
df_kurt = df1.kurt().abs()
df_skew = df1.skew().abs()
```

```python
df_kurt_dict = df_kurt.to_dict()
df_skew_dict = df_skew.to_dict()
```

```python
qualifiedkurt = {k:v for k, v in df_kurt_dict.items() if v <= 3}
qualifiedskew = {k:v for k, v in df_skew_dict.items() if v <= 0.8}
final_dict = {k: qualifiedkurt[k] + qualifiedskew[k] for k in qualifiedkurt if k in qualifiedskew}
#for keys, values in qualifiedkurt.items():
 #   print(keys, ':', values)
```
